```julia
function hpfilter(y, λ::Real)
    # Convert y to vector
    if !isa(y, Vector)
        try
            y = convert(Vector,y)
        catch
            error("Series must be convertible to Vector")
        end
    end

    # Indices of consecutive NaN elements at beginning
    i = 1
    j = length(y)
    while isnan(y[i])
        i = i+1
    end
    while isnan(y[j])
        j = j-1
    end

    # Filter and adjust for NaNs
    yt_, yf_ = hpfilter_(y[i:j], λ)
    yt = [fill(NaN, i-1); yt_; fill(NaN, length(y)-j)]
    yf = [fill(NaN, i-1); yf_; fill(NaN, length(y)-j)]

    return yt, yf
end

function hpfilter_{T<:Real}(y::Vector{T}, λ::Real)
    n = length(y)
    a = spzeros(n,n)
    for i = 3:n-2
        a[i,i]   = 6λ+1
        a[i,i+1] = -4λ
        a[i,i+2] = λ
        a[i,i-2] = λ
        a[i,i-1] = -4λ
    end

    a[2,2] = 1+5λ
    a[2,3] = -4λ
    a[2,4] = λ
    a[2,1] = -2λ
    a[1,1] = 1+λ
    a[1,2] = -2λ
    a[1,3] = λ

    a[n-1,n-1] = 1+5λ
    a[n-1,n-2] = -4λ
    a[n-1,n-3] = λ
    a[n-1,n]   = -2λ
    a[n,n]     = 1+λ
    a[n,n-1]   = -2λ
    a[n,n-2]   = λ

    yt = a\y
    yf = y-yt

    return yt, yf
    end
```

```julia
gdp_raw = get_data(f, "GDPC1",frequency="q",observation_start="1948-01-01");
gdpdef_raw = get_data(f, "GDPDEF",frequency="q",observation_start="1948-01-01");
pcesv_raw = get_data(f, "PCESV",frequency="q",observation_start="1948-01-01");
pcnd_raw = get_data(f, "PCND",frequency="q",observation_start="1948-01-01");
fpi_raw = get_data(f, "FPI",frequency="q",observation_start="1948-01-01");
pop_raw = get_data(f, "CNP16OV",frequency="q",observation_start="1948-01-01",observation_end="2017-10-01");
hours_raw = get_data(f, "HOANBS",frequency="q",observation_start="1948-01-01");
w_raw = get_data(f, "COMPRNFB",frequency="q",observation_start="1948-01-01");
```

```julia
y = convert(Array,(gdp_raw.df[4]*10^9)./(pop_raw.df[4]*1000));
cesv = convert(Array,((pcesv_raw.df[4]./(gdpdef_raw.df[4]/100))*10^9)./(pop_raw.df[4]*1000));
cnd = convert(Array,((pcnd_raw.df[4]./(gdpdef_raw.df[4]/100))*10^9)./(pop_raw.df[4]*1000));
c = cesv+cnd;
inv = convert(Array,(fpi_raw.df[4]./(gdpdef_raw.df[4]/100)));
hours = convert(Array,hours_raw.df[4]./(pop_raw.df[4]*1000));
w = convert(Array,w_raw.df[4]);
```

```julia
data=DataFrame();
data[:y]=y;
data[:c]=c;
data[:inv]=inv;
data[:n]=hours;
data[:w]=w;
```

```julia
logdata=DataFrame();
logdata[:y]=log(data[:y]);
logdata[:c]=log(data[:c]);
logdata[:inv]=log(data[:inv]);
logdata[:n]=log(data[:n]);
logdata[:w]=log(data[:w]);
```

```julia
hpdata=DataFrame();
hpdata[:y]=hpfilter(logdata[:y],1600)[2];
hpdata[:c]=hpfilter(logdata[:c],1600)[2];
hpdata[:inv]=hpfilter(logdata[:inv],1600)[2];
hpdata[:n]=hpfilter(logdata[:n],1600)[2];
hpdata[:w]=hpfilter(logdata[:w],1600)[2];
```
